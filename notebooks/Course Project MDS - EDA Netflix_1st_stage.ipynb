## Оценивание и штрафы



Групповой проект оценивается исходя из устной защиты (презентации) командой выполненного ими проекта перед комиссией. Комиссия составляется из преподавателей курса, ассистентов курса, а также приглашенных экспертов из индустрии.

В рамках каждой защиты комиссия выставляет два типа оценок.

Первая оценка — это оценка для всего проекта в целом. Она ставится комиссией, исходя из глубины проработки командой предметной области, сложности применяемых в проекте инструментов, а также прочих аспектов, связанных с содержанием выполняемого в проекте ТЗ. При выставлении этой оценки комиссия ориентируется на критерии, более подробно представленные ниже в разделе "Критерии оценки проекта".

Вторая оценка — это индивидуальная оценка каждого участника команды. Именно эта оценка выставляется впоследствии в ведомость. Индивидуальная оценка участника команды определяется комиссией, исходя из того, насколько участник ориентируется в выполненном командой проекте, как он отвечает на адресованные персонально ему вопросы комиссии, насколько в принципе прослеживается его роль в проекте, а также исходя из других аспектов, связанных с участием конкретного человека в деятельности команды. Индивидуальная оценка не может превосходить оценку команды за проект в целом (первую оценку), однако, в случае когда участник спокойно и корректно отвечает на все адресованные ему вопросы, свободно ориентируется в материале и содержании выполненного проекта, а также когда его роль в работе команды не вызывает сомнений, — в таком случае, индивидуальная оценка участника становится равной оценке для всего проекта (т.е. в идеальном случае — вторая оценка равна первой оценке).

Во избежание субъективности и/или предвзятости оценки тех или иных членов комиссии в отношении тех или иных команд, а также каких-либо иных проявлений неравных условий — распределение команд между комиссиями осуществляется абсолютно рандомно с использованием средств библиотеки Numpy :)) — в качестве рэндом сида выступает дата, в которую производится распределение, например, `np.random.seed(11032026)`.

Для сохранения справедливости и беспристрастности случайного распределения, любые переводы команд между комиссиями (смена комиссии) не допускаются. Присутствие студента на защите проекта является обязательным; вклад студента при его отсутствии не оценивается и автоматически считается равным нулю. #### **Возрастные рейтинги**

В данных представленно 13 возрастных рейтингов групп, с помощью поиска в гугле можно разделить данные на 2 группы:

**1. Группа рейтингов для телевидения:**

* TV-14 - детям до 14 лет просмотр не рекомендуется
* TV-PG - рекомендуется родительский контроль
* TV-MA - программа предназначена исключительно для взрослых и может содержать контент, не подходящий для детей младше 17 лет
* TV-Y - программа предназначена для детей всех возрастов
* TV-Y7 - программа предназначена для детей в возрасте от 7 лет и старше
* TV-G - программа предназначена для широкой зрительской аудитории и подходит для всех возрастов
* TV-Y7-FV - программа ориентирована на детей младшего школьного возраста

**2. Группа рейтингов для фильмов:**

* PG-13 - просмотр нежелателен для детей до 13 лет
* R - фильм предназначен для взрослой аудитории
* NR - фильм не проходил официальную оценку и сертификацию 
* UR - фильм не проходил официальную оценку и сертификацию 
* PG - фильм или шоу рекомендуется смотреть в присутствии родителей
* G - фильм не имеет возрастных ограничений и допущен для просмотра зрителям всех возрастов

Рассмотрим категории сериалов и кино раздельно, возможно там будут неочевидные тренды. Добавим новый признак **'tv'**, где значение '1'- сериал, '0'- фильм/мультфильм

Так же будет целесообразно добавить в датасет еще один новый признак **age** и разделим категории рейтингов по возрастным группам, так сказать по ЦА:

1. Первая группа 'kid' - фильмы и программы для детей (TV-Y, TV-Y7, TV-G, TV-Y7-FV, G)
2. Вторая группа 'teen'- фильмы и программы для подростков (TV-14, TV-PG, PG-13, PG)
3. Третья группа 'adult' - фильмы и программы для взрослых (TV-MA, R, NR, UR)

Так как тут присутствуют фильмы/сериалы которые предназначены для просмотра с родителями или в семейном кругу, добавим еще один бинарный признак **'family_friendly'**, где '1' - контент предназначенный для просмотра с семьей, а '0' - контент для взрослых:

* '1' - TV-Y, TV-Y7, TV-G, TV-Y7-FV, G, TV-14, TV-PG, PG-13, PG
* '0' - TV-MA, R, NR, UR

```python
data_clean['tv'] = data_clean['rating'].apply(lambda x: 1 if x in ['TV-14', 'TV-PG', 'TV-MA', 'TV-Y','TV-Y7', 'TV-G', 'TV-Y7-FV'] else 0)
data_clean['family_friendly'] = data_clean['rating'].apply(lambda x: 1 if x in ['TV-Y', 'TV-Y7', 'TV-G', 'TV-Y7-FV','G', 'TV-14', 'TV-PG', 'PG-13', 'PG'] else 0)

data_clean['age'] = data_clean['rating'].apply(lambda x: 'kid' if x in ['TV-Y', 'TV-Y7', 'TV-G', 'TV-Y7-FV','G'] 
                                else ('teen' if x in ['TV-14', 'TV-PG', 'PG-13', 'PG']
                                else 'adult' ))

data_clean.head(10)
```

Кино и сериалы будем анализировать отдельно, так как у них могут отличасться признаки, факторы успеха, аудитория и т.д.

```python
# до разделения заполним пропуски
# год выпуска = 0, т.к. информации нет, но null может помешать
imdb['startYear'] = imdb['startYear'].fillna(0).astype(int)

# для отсутствующих жанров установим "неизвестно"
imdb['genres'] = imdb['genres'].fillna('unknown')

# уберем строки с отсутствующими названиями, колонку с названием на ориг. языке - не будем использовать
imdb = imdb[imdb['primaryTitle'].notna()].drop(columns='originalTitle').reset_index(drop=True)
```

```python
# фильтруем данные и сразу разделяем кино и сериалы
imdb_movies = imdb[(imdb['titleType'].isin(['movie','tvMovie'])) & (imdb['startYear']>=1940) & (imdb['startYear']<=2017)]
imdb_shows = imdb[(imdb['titleType'].isin(['tvMiniSeries','tvSeries'])) & (imdb['startYear']>=1940) & (imdb['startYear']<=2017)]
```

```python
# из начальных данных оставляем только фильмы, фильтруя по созданному нами признаку
movies = data_clean[data_clean['tv']==0]
movies.shape[0]
```

```python
# функция для нормализации названий - превращает все в lowercase, убирает символы, оставляет только один пробел
def normalize(x):
    x = str(x)
    x = x.lower()
    x = x.replace('&',' and ')
    x = x.replace('–','-')
    x = x.replace('—','-')
    x = re.sub(r'[^a-z0-9\s]', ' ', x)
    x = re.sub(r'\s+', ' ', x)

    return x.strip()

# копии чтобы все работало корректно
movies = movies.copy()
imdb_movies = imdb_movies.copy()

# примение нормализации названий
movies['norm_title'] = movies['title'].apply(normalize)
imdb_movies['norm_title'] = imdb_movies['primaryTitle'].apply(normalize)

# объединение данных только по совпадающим названиям
movies_enriched = pd.merge(movies, imdb_movies, how='inner', on= 'norm_title')
```

```python
# фильтрация по году выпуска фильма, так как встерчаются разные фильмы с одинаковыми названиями
movies_enriched = movies_enriched[(movies_enriched['startYear'] == movies_enriched['release year'])]

# если было несколько совпадений - убираем дубликаты по названию и году выпуска
movies_enriched = movies_enriched.drop_duplicates(subset=['title', 'release year'], keep='first').reset_index(drop=True)

print(f"Строк: {movies_enriched.shape[0]}")
print(f"Уникальных фильмов: {movies_enriched['title'].nunique()}")
print(f"Уменьшилось фильмов на {len(movies) - len(movies_enriched)}")
```

```python
# наши сериалы
tv = data_clean[data_clean['tv']==1]
tv.shape
```

```python
# нормализация названий и объединение данных
tv = tv.copy()
imdb_shows = imdb_shows.copy()

tv['norm_title'] = tv['title'].apply(normalize)
imdb_shows['norm_title'] = imdb_shows['primaryTitle'].apply(normalize)

tv_enriched = pd.merge(tv, imdb_shows, how='inner', on= 'norm_title')
```

```python
# учитываем то, что в наших данных не всегда точно указывается год выпуска сериалов, иногда это год выпуска того сезона, который добавлен на нетфликс
# берем окошко в 15 лет - чтобы не объединенять оргиналы с ремейками, при этом включить старые и длинные сериалы
tv_enriched = tv_enriched[(tv_enriched['startYear'] <= tv_enriched['release year']) & (tv_enriched['startYear'] >= tv_enriched['release year'] - 15)]

# убираем дубли и случаи неправильных объединений для ремейков и прочих сериалов с одинаковыми названиями
# вычисляем разницу года выпуска нетфликса и imdb
tv_enriched['year_diff'] = tv_enriched['release year'] - tv_enriched['startYear']

# датасет сортируется по возрастанию разницы годов выпуска, удаляется все, кроме самой меньшей разницы - ближайший метч
tv_enriched = (tv_enriched.sort_values('year_diff').drop_duplicates(subset=['title', 'release year'], keep='first'))
tv_enriched = tv_enriched.drop(columns='year_diff').reset_index(drop=True)

print(f"Строк: {tv_enriched.shape[0]}")
print(f"Уникальных шоу: {tv_enriched['title'].nunique()}")
print(f"Уменьшилось сериалов на {len(movies) - len(movies_enriched)}")
```
